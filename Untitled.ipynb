```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('data.csv')
x = data.iloc[:, 0].values.reshape(-1, 1) 
y = data.iloc[:, 1].values.reshape(-1, 1)
lr = LinearRegression()
lr.fit(x, y)
y_pred = lr.predict(x)
```

```python
plt.scatter(x, y)
plt.plot(x, y_pred, color='red')
plt.show()
```
